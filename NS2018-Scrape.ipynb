```python
from bs4 import BeautifulSoup
import requests
import re

r  = requests.get("https://www.conftool.pro/dh2018/index.php?page=browseSessions&print=head&doprint=yes&presentations=show")
soup = BeautifulSoup(r.content, "lxml")
print(soup.prettify())
```
